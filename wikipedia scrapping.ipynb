```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page.content
```

```python
content = BeautifulSoup(page.content,'html.parser')
```

```python
print(content.prettify())
```

```python
Header_Tags=[]
```

```python
Tags = content.find_all('h2' ,class_='mp-h2')
Tags[0:]
```

```python
for i in Tags:
    Header_Tags.append(i.text)
Header_Tags[0:]
```

```python
import pandas as pd
Header_tags=pd.DataFrame({})
Header_tags['Tags']=Header_Tags
```
