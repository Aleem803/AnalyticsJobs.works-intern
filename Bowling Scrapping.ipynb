```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
page.content
```

```python
BS = BeautifulSoup(page.content,'html.parser')
print(BS.prettify())
```

```python
Ranking = []
Player = []
Team = []
Rating = []
Carrier_best_Rating = []
```

```python
pos = BS.find_all('td',class_ = 'rankings-block__position')
for i in pos:
    Ranking.append(i.text)
Ranking    
```

```python
name=BS.find_all('td',class_ ='rankings-block__top-player-container')
for i in name:
    Player.append(i.text)
Player
```

```python
country = BS.find_all('div',class_='rankings-block__banner--nationality')
for i in country:
    Team.append(i.text)
Team
```

```python
score = BS.find_all('div',class_ = 'rankings-block__banner--rating')
for i in score:
    Rating.append(i.text)
Rating
```

```python
best = BS.find_all('span',class_ = 'rankings-block__career-best-text')
for i in best:
    Carrier_best_Rating.append(i.text)
Carrier_best_Rating
```

```python
position = BS.find_all('td',class_ = 'table-body__cell table-body__cell--position u-text-right')
for i in position:
    Ranking.append(i.text)
Ranking
```

```python
Name = BS.find_all('td',class_ = 'table-body__cell rankings-table__name name')
for i in Name:
    Player.append(i.text)
Player
```

```python
Country = BS.find_all('span',class_='table-body__logo-text')
for i in Country:
    Team.append(i.text)
Team
```

```python
Score = BS.find_all('td',class_ = 'table-body__cell rating')
for i in Score:
    Rating.append(i.text)
Rating
```

```python
Best = BS.find_all('td',class_ = 'table-body__cell u-text-right u-hide-phablet')
for i in Best:
    Carrier_best_Rating.append(i.text)
Carrier_best_Rating
```

```python
import pandas as pd
Bowling= pd.DataFrame({})
Bowling['position'] = Ranking
Bowling['Name'] = Player
Bowling['Country'] = Team
Bowling['Score'] = Rating
Bowling['Best'] = Carrier_best_Rating
Bowling
```
