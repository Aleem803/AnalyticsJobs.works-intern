```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
page.content
```

```python
BS = BeautifulSoup(page.content,'html.parser')
print(BS.prettify())
```

```python
Position = []
Player = []
Team = []
Rating = []
```

```python
pos = BS.find_all('div', class_ = 'rankings-block__banner--pos')
pos
```

```python
for i in pos:
    Position.append(i.text)
Position
```

```python
name = BS.find_all('div', class_ = 'rankings-block__banner--name')
name
```

```python
for i in name:
    Player.append(i.text)
Player
```

```python
country = BS.find_all('span',class_ = 'rankings-block__banner--nation')
country
```

```python
for i in country:
    Team.append(i.text)
Team
```

```python
rating = BS.find_all('div', class_ = 'rankings-block__banner--rating')
rating
```

```python
for i in rating:
    Rating.append(i.text)
Rating
```

```python
pos1 = BS.find_all('td', class_ ='table-body__cell table-body__cell--position u-text-right')
pos1
```

```python
for i in pos1:
    Position.append(i.text)
Position
```

```python
name1 = BS.find_all('td',class_ = 'table-body__cell name')
name1
```

```python
for i in name1:
    Player.append(i.text)
Player
```

```python
nation = BS.find_all('span', class_ = 'table-body__logo-text')
nation
```

```python
for i in nation:
    Team.append(i.text)
Team
```

```python
rating1 = BS.find_all('td', class_ = 'table-body__cell u-text-right rating')
rating1
```

```python
for i in rating1:
    Rating.append(i.text)
Rating
```

```python
len(Position),len(Player),len(Team),len(Rating)
```

```python
import pandas as pd
Ranking = pd.DataFrame({})
Ranking['pos1'] = Position
Ranking['name1'] = Player
Ranking['nation'] = Team
Ranking['rating1'] = Rating
```
